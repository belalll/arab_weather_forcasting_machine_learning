```python
import numpy as np
import pandas as pd
```

```python
d = pd.read_csv("daily_weather_2020.csv")
d
```

```python
train=d[d['Country/Region'].isin(['Egypt','Algeria','Sudan','Saudi Arabia','Tunisia','Somalia','Morocco','Iraq','Jordan','Qatar','Iraq','Oman'])]
train
```

```python
train["summary"].unique()
```

```python
new_data = {  "Mostly cloudy throughout the day.":"Mostly cloudy throughout the day.",
            "Partly cloudy throughout the day.":"Partly cloudy throughout the day.",
            "Humid and partly cloudy throughout the day.":"Humid and partly cloudy throughout the day.",
            "Humid and mostly cloudy throughout the day.":"Humid and mostly cloudy throughout the day."}


new_data1 = {"Rain in the afternoon and evening.":"Rain in the afternoon and evening.",
             "Possible light rain in the evening.":"Possible light rain in the evening.",
             "Possible light rain overnight.":"Possible light rain overnight.",
             "Light rain starting in the afternoon.":"Light rain starting in the afternoon.",
             "Light rain throughout the day.":"Light rain throughout the day.",
             "Rain in the evening and overnight.":"Rain in the evening and overnight.",
             "Possible drizzle overnight.":"Possible drizzle overnight.",
             "Light rain in the morning.":"Light rain in the morning.",
             "Possible light rain in the morning.":"Possible light rain in the morning.",
             "Possible drizzle in the morning.":"Possible drizzle in the morning.",
             "Possible light rain in the morning and afternoon.":"Possible light rain in the morning and afternoon.",
             "Light rain in the morning and afternoon.":"Light rain in the morning and afternoon.",
             "Rain overnight.":"Rain overnight.",
             "Light rain in the evening and overnight.":"Light rain in the evening and overnight.",
             "Light rain until evening.":"Light rain until evening.",
             "Possible light rain until evening.":"Possible light rain until evening.",
             "Humid throughout the day.":"Humid throughout the day.",
             "Drizzle in the morning.":"Drizzle in the morning.",
             "Possible light rain in the morning and overnight.":"Possible light rain in the morning and overnight.",
             "Possible light rain in the evening and overnight.":"Possible light rain in the evening and overnight.",
             "Light rain until morning, starting again in the evening.":"Light rain until morning, starting again in the evening.",
             "Rain until morning, starting again in the evening":"Rain until morning, starting again in the evening",
             "Rain until afternoon, starting again overnight.":"Rain until afternoon, starting again overnight.",
             "Rain and humid throughout the day.":"Rain and humid throughout the day.",
             "Light rain and windy in the morning.":"Light rain and windy in the morning.",
             "Possible light rain in the afternoon and evening.":"Possible light rain in the afternoon and evening.",
             "Light rain overnight.":"Light rain overnight." 
            }


new_data2 = { "Clear throughout the day.":"Clear throughout the day." }

new_data3 = { "Overcast throughout the day.":"Overcast throughout the day.",
             "Humid and overcast throughout the day.":"Humid and overcast throughout the day." }



new_data4 = { "Dangerously windy overnight.":"Dangerously windy overnight.",
             "Windy starting in the afternoon.":"Windy starting in the afternoon.",
             "Windy until evening.":"Windy until evening.",
             "Windy in the morning.":"Windy in the morning.",
             "Windy overnight.":"Windy overnight.",
             "Windy in the afternoon.":"Windy in the afternoon.",
             "Dangerously windy in the evening and overnight.":"Dangerously windy in the evening and overnight.",
             "Windy until morning, starting again in the evening.":"Windy until morning, starting again in the evening.",
             "Dangerously windy in the morning.":"Dangerously windy in the morning.",
             "Windy in the evening.":"Windy in the evening.",
             "Windy in the evening and overnight.":"Windy in the evening and overnight.",
             "Windy in the morning and afternoon.":"Windy in the morning and afternoon.",
             "Dangerously windy starting in the afternoon.":"Dangerously windy starting in the afternoon."
            }


train["cloudness"] = train["summary"].map(new_data)
train["raining"] = train["summary"].map(new_data1)
train["clearing"] = train["summary"].map(new_data2)
train["overcasting"] = train["summary"].map(new_data3)
train["winding"] = train["summary"].map(new_data4)
```

```python
train
```

```python
train['DateTime'] = pd.to_datetime(train['time'])
from datetime import date
train['year']= train['DateTime'].dt.year
train['month']= train['DateTime'].dt.month
train['day']= train['DateTime'].dt.day
train
```

```python
train.drop(columns="precipAccumulation",inplace=True)
train.drop(columns="precipType",inplace=True)
train.drop(columns="Province/State",inplace=True)
train.drop(columns="DateTime",inplace=True)
train.drop(columns="time",inplace=True)
train.drop(columns="summary",inplace=True)
train.drop(columns="icon",inplace=True)
train.drop(columns="sunriseTime",inplace=True)
train.drop(columns="sunsetTime",inplace=True)
train.drop(columns="precipIntensityMaxTime",inplace=True)
train.drop(columns="precipIntensityMax",inplace=True)
train.drop(columns="precipIntensity",inplace=True)
train.drop(columns="temperatureHighTime",inplace=True)
train.drop(columns="temperatureLowTime",inplace=True)
train.drop(columns="apparentTemperatureHigh",inplace=True)
train.drop(columns="apparentTemperatureHighTime",inplace=True)
train.drop(columns="apparentTemperatureLow",inplace=True)
train.drop(columns="apparentTemperatureLowTime",inplace=True)
train.drop(columns="windGustTime",inplace=True)
train.drop(columns="uvIndexTime",inplace=True)
train.drop(columns="temperatureMinTime",inplace=True)
train.drop(columns="temperatureMaxTime",inplace=True)
train.drop(columns="apparentTemperatureMinTime",inplace=True)
train.drop(columns="apparentTemperatureMaxTime",inplace=True)
train.drop(columns="apparentTemperatureMin",inplace=True)
train.drop(columns="apparentTemperatureMax",inplace=True)
train.drop(columns="year",inplace=True)
train.drop(columns="Unnamed: 0",inplace=True)

```

```python
train
```

```python
train['cloudness'].fillna(value="not cloudy day",inplace=True)
train['clearing'].fillna(value="not clear day",inplace=True)
train['overcasting'].fillna(value="not overcast day",inplace=True)
train['winding'].fillna(value="not wind day",inplace=True)
train['raining'].fillna(value="not rainy day",inplace=True)
train
```

```python
fisrt__df = train.drop(['clearing','overcasting','winding','cloudness'],axis=1)
fisrt__df.head(5)
```

```python
fisrt__df['raining'].unique()
```

```python
def conert_column(df,c):
    n = df[c].unique().shape[0]
    for i in range(0,n):
        df[c] = df[c].str.replace(df[c].unique()[i],str(i))
    df[c] = pd.to_numeric(df[c])
    
```

```python
conert_column(fisrt__df,'raining')
```

```python
fisrt__df['raining'].unique()
```

```python
fisrt__df.head(5)
```

```python
fisrt__df['raining'] = pd.to_numeric(fisrt__df['raining'])
```

```python
qualitative =  [f for f in fisrt__df.columns if fisrt__df.dtypes[f] == 'object']
qualitative
```

```python
for i in qualitative :
    
    fisrt__df[i] = pd.Categorical(fisrt__df[i])
    fisrt__df[i] = pd.get_dummies(fisrt__df[i])
```

```python
from sklearn.model_selection import train_test_split
#train['summary'] = np.log1p(train['summary'])


train_df, val_df = train_test_split(fisrt__df, test_size=0.05, random_state=80)

X_train = train_df.drop(columns=['raining'])
y_train = train_df['raining']

X_val = val_df.drop(columns=['raining'])
y_val = val_df['raining']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(40,40))
cor = fisrt__df.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()
```

```python
# Consider correlations only with the target variable
cor_target = abs(cor['raining'])

#Select correlations with a correlation above a threshold
features2 = cor_target[cor_target>0.2]

feat=list(features2.keys())
feat.remove('raining') 

print(feat)
new_X = X_train[feat]
new_X_val = X_val[feat]
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2 , f_regression

KBest = SelectKBest(score_func = f_regression, k = 15)
KBest = KBest.fit(X_train,y_train)

df_scores = pd.DataFrame({'features': X_train.columns, 'f_regression': KBest.scores_, 'pValue': KBest.pvalues_ })
df_scores
```

```python
cols = KBest.get_support(indices=True)
cols
features = X_train.columns[cols]
list(features)
```

```python
final_train = X_train[features]
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import chi2
from sklearn.linear_model import LogisticRegression

poly_reg_model=PolynomialFeatures(degree=1)
X_poly=poly_reg_model.fit_transform(final_train)
# poly_reg.fit(X_poly,y)
lin_reg_model=LinearRegression()
lin_reg_model.fit(X_poly,y_train)
```

```python
model=LogisticRegression(solver='liblinear')
model.fit(X_poly,y_train)
```

```python
final_val = X_val[features]
X_val_poly=poly_reg_model.fit_transform(final_val )
print("The accuracy of the classifier on the validation set is ", (model.score(X_val_poly,y_val)))
```

```python
from sklearn import svm
sv_model=svm.SVC()
sv_model.fit(X_poly,y_train)
sv_model.score(X_val_poly,y_val)
```
